```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

二壘有人：

```python
# Single
n_bh = 100
p_bh = 0.139 #50% chance to go home
x_bh = np.arange(0,101)
pmf_bh = binom.pmf(x_bh, n=n_bh, p=p_bh)
# Home Run
n_hr = 100
p_hr = 0.101
x_hr = np.arange(0,101)
pmf_hr = binom.pmf(x_hr, n=n_hr, p=p_hr)
```

```python
plt.figure(figsize=(16,12))
plt.stem((x_bh), pmf_bh, linefmt='b-', markerfmt='bo', basefmt=" ", label='Single')
plt.stem((x_hr)*2, pmf_hr, linefmt='r-', markerfmt='ro', basefmt=" ", label='Home Run')
plt.title('Runner on Second Base')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.legend()
plt.show()
low_s, high_s = binom.interval(0.9, n_bh, p_bh)
low_hr, high_hr = binom.interval(0.9, n_hr, p_hr)
print(f"Single 95% CI: {low_s}, {high_s}")
print(f"Home Run 95% CI: {low_hr*2}, {high_hr*2}")
```

二三壘有人：

```python
# Single
n_bh = 100
p_bh = 0.279
n_sim = 100000  # 模擬次數
results = []
for _ in range(n_sim):
    # 安打次數
    H = np.random.binomial(n_bh, p_bh)
    # 二壘跑者得分次數
    R = np.random.binomial(H, 0.5)
    # 總得分
    total_runs = H + R
    results.append(total_runs)
# 畫 PMF（用相對頻率近似）
values, counts = np.unique(results, return_counts=True)# 得分,次數
pmf_bh = counts / n_sim
x_full = np.arange(0, 201)
pmf_bh_full = np.zeros_like(x_full, dtype=float)
pmf_bh_full[values] = pmf_bh
# Home Run
n_hr = 100
p_hr = 0.101
x_hr = np.arange(0,101)
pmf_hr = binom.pmf(x_hr, n=n_hr, p=p_hr)
```

```python
plt.figure(figsize=(16,12))
plt.stem(x_full, pmf_bh_full, linefmt='b-', markerfmt='bo', basefmt=" ", label='Single')# x,y軸
plt.stem((x_hr)*3, pmf_hr, linefmt='r-', markerfmt='ro', basefmt=" ", label='Home Run')
plt.title('Runners on Second and Third')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
